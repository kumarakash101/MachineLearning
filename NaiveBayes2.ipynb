```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
target = df['spam']
inputs = df['Message']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = ["Hi jim i just wanted to say thank you for your help last week",
         "Winner!! you have won a $100 amazon gift card click here to redeem"]

emails_count= v.transform(emails)
model.predict(emails_count)

# 1= yes its a spam
# 0=not a spam
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count,y_test)
```

```python
# The model is super effective in identifying the spam
```

```python
# Now you might have noticed that to check any email we have to convert it into numbers by using transform() everytime
# this is inconvient to sklearn provies us a tool called pipeline that will do this task for you
# Pipline can be used to transform data as programmed by the user
# like one process after the another
# Pipeline` allows you to sequentially apply a list of transformers to preprocess the data and, if desired, conclude the sequence with a final
# :term:`predictor` for predictive modeling
```

```python
 from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```

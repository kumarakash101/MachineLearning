```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing digits dataset

from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# Lets save the digits dataset into pandas dataframe

df = pd.DataFrame(digits.data)
```

```python
# Now lets append the target column in our main dataframe

df['target'] = digits.target
df.head()
```

```python
# Dividing the feature and target columns

X = df.drop('target',axis=1)
y = df.target
```

```python
# Now lets do the train test split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Seeing how the distribution took place

print("X train lenght",len(X_train))
print("X test lenght",len(X_test))
print("y train lenght",len(y_train))
print("y test lenght",len(y_test))
```

```python
# As we can see that the test train split happened perfectly

test=int((20/100)*1797)
test
```

```python
# Now lets import random forest classifier to train our model

from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Now lets see what parameters our model has used to train itself 
# Here as we can see that n_estimators=100
# which means that the random forest used 100 decision trees to train itself

model.get_params()
```

```python
model.score(X_test,y_test)
```

```python
# Now lets plot a confusion matrix to see how our model performed(i.e. where it performed ok or where it made an error)

y_pred = model.predict(X_test)
y_pred
```

```python
# now once you have the y predicted its a good practice to visualize the prediction of our model using confusion matrix
# We need actual answers which are stored in y_test and then we need y_pred which is the values predicted by the model
# and now we will plot both the values in the graph to see how the prediction looks
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# A confusion matrix is a table used to evaluate the performance of a classification model in machine learning. 
# It compares the model's predicted outcomes with the actual, known values of a dataset, showing how well the model is predicting different classes. 
# Essentially, it helps visualize where the model is making errors and where it's performing well. 

cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# Tip:- you should use seaborn when you have two set of values which are forming a matrix 
# They help us in visulizing relationship, pattern or magnitude
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(5,5))

sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

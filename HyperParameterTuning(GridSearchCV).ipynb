```python
# As we can see that c2 has the best performance that means the hyperpatameter of c2 is the best but this is still not efficient
# becuase we have to provide all the values manually to the system
# one other manual approach is that we can run a for loop which will be a littlebit convenient
```

```python
mykernals = ['linear','rbf']
myc = [1,10,20,30]

avg_score = {}
# empty dictionary used to save key-value pair seperated by :

for k in mykernals:
    for c in myc:
        c1=cross_val_score (SVC(kernel=k,C=c,gamma='auto'),iris.data, iris.target, cv=5)
        avg_score[k+'-'+str(c)]= np.mean(c1)
```

```python
# So as we can see that this process is little bit less manual but it still requires quite a bit of manual labour.
# Also this method is pretty harsh on the system resources imagin if our dataset had 1M data points then this would have taken a lot of time\

avg_score
```

```python
# Now to avoide all the overhead issues like manual intervention and system overloading sklearn offers an API callled as
# GridSearchCv that will do all this woek in one line and in a more effective way

from sklearn.model_selection import GridSearchCV
```

```python
classifier.fit(iris.data,iris.target)
```

```python
# Now with that being said we also face the problem of which ml algorith to use
# That is we need to find the optimal algo. which gives us the best performance
# so we use grid serach cv as we but this time instead of hyperparameter we use different models
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params ={
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    }
}
```

```python
classifier = GridSearchCV(SVC(gamma='auto'),{
    # this is the parameter grid
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5, return_train_score=False
                         )          
```

```python
scores=[]

for model_names, mparams in model_params.items():
    classi = GridSearchCV(mparams['model'],mparams['params'],cv=5, return_train_score=False)
    classi.fit(iris.data, iris.target)
    scores.append({
        'model':model_names,
        'best_score':classi.best_score_,
        'best_params':classi.best_params_
    })
```

```python
scores
```

```python
scores_df = pd.DataFrame(data=scores, columns=['model','best_score','best_params'])
```
